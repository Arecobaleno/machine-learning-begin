```python
from skimage import io
from sklearn.cluster import KMeans
import numpy as np
```

```python
image = io.imread('example.png')
image.shape
```

```python
rows = image.shape[0]
cols = image.shape[1]

image = image.reshape(rows*cols,4)
```

```python
kmeans = KMeans(n_clusters=128)
kmeans.fit(image)
```

```python
labels = np.asarray(kmeans.labels_)
```

```python
labels = labels.reshape(rows,cols)
```

```python
io.imshow(labels)
```
