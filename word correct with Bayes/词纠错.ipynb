```python
import re
import collections
```

```python
text = open('word.txt').read()
```

```python
def Word(text):
    return re.findall('[a-z]+',text.lower())

word = Word(text)
```

```python
def transform(word):
    model = collections.defaultdict(lambda : 1)
    for f in word:
        model[f]+=1
    return model

model = transform(word)
```

```python
model
```

```python
alphabet = set('abcdefghijklmnopqrstuvwxyz')
#编辑距离为1
def edit1(word):
    n = len(word)
    return set([word[0:i] + word[i+1:] for i in range(n)]+
               [word[0:i] + c + word[i:] for i in range(n) for c in alphabet]+
               [word[0:i] + word[i+1] + word[i] + word[i+2:] for i in range(n-1)]+
               [word[0:i] + c + word[i+1:] for i in range(n) for c in alphabet])

def edit2(word):
    return set(e2  for e1 in edit1(word) for e2 in edit1(e1))

def known(words):
    return set(w for w in words if w in model)
```

```python
def correct(word):
    candidates = known([word]) or known(edit1(word)) or known(edit2(word)) or [word]
    return max(candidates,key=lambda w:model[w])
```
