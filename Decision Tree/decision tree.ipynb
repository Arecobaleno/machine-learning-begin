```python
import pandas as pd
```

```python
iris_data = pd.read_csv('iris.data.csv')
```

```python
iris_data.columns = ['sepal_length_cm','sepal_width_cm','petal_length_cm','petal_width_cm','kind']
```

```python
iris_data.describe()
```

```python
import seaborn as sns
```

```python
sns.pairplot(iris_data.dropna(),hue = 'kind')#dropna 
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
for index,column in enumerate(iris_data.columns):
    if column == 'kind':
        continue
    plt.subplot(2,2,index+1)
    sns.violinplot(x = 'kind', y = column, data = iris_data)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

all_input = iris_data[['sepal_length_cm','sepal_width_cm','petal_length_cm','petal_width_cm']].values
all_class = iris_data['kind'].values

train_inputs,test_input,train_label,test_label = train_test_split(all_input,all_class,train_size = 0.75,random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier()
tree.fit(train_inputs,train_label)
predict_label = tree.predict(test_input)
```

```python
import sklearn.metrics
sklearn.metrics.accuracy_score(test_label,predict_label)
```
